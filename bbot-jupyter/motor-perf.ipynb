# Motor Performance
The purpose of this notebook is to explore the performance of the stepper motor. We will be adjusting the torque/speed ratio and plotting rotational reading over time.

```python
import numpy as np
import matplotlib.pyplot as plt
import json
import math
import random
```

```python
def load_data(path, perc=1.0):
    with open(path) as f:
        data = json.load(f)
    arr = np.asarray(data, dtype=np.float32)
    time = arr[:,0]
    avg_rotation = arr[:,1]
    raw_rotation = arr[:,2]
    x = arr[:,3]
    y = arr[:,4]
    z = arr[:,5]
    
    return (time, avg_rotation, raw_rotation, x, y, z)
```

```python

time, avg_rotation, raw_rotation, x, y, z = load_data("motor/readings.json")

print(np.average(avg_rotation))

plt.plot(y, color='blue')
plt.plot(x, color='yellow')
plt.plot(avg_rotation, color='red')
plt.plot(z, color='green')

plt.show()


```
